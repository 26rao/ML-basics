```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
st=StandardScaler()
test['RoomService']=st.fit_transform(test[['RoomService']])
test['FoodCourt']=st.fit_transform(test[['FoodCourt']])
test['VRDeck']=st.fit_transform(test[['VRDeck']])
test['Age']=st.fit_transform(test[['Age']])
```

```python
from sklearn.model_selection import train_test_split
x=train.drop(columns='Transported')
y=train['Transported']
X_test=test.copy()
```

```python
x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
mod_rf=RandomForestClassifier()
mod_dc=DecisionTreeClassifier()
mod_lr=LogisticRegression()
mod_kn=KNeighborsClassifier()
```

```python
model_rf=mod_rf.fit(x_train,y_train)
rf_pred=model_rf.predict(x_val)
acc_rf=accuracy_score(y_val,rf_pred)
print(acc_rf)
```

```python
model_dc=mod_dc.fit(x_train,y_train)
dc_pred=model_dc.predict(x_val)
acc_dc=accuracy_score(y_val,dc_pred)
print(acc_dc)
```

```python
model_lr=mod_lr.fit(x_train,y_train)
lr_pred=model_lr.predict(x_val)
acc_lr=accuracy_score(y_val,lr_pred)
print(acc_lr)
```

```python
model_kn=mod_kn.fit(x_train,y_train)
kn_pred=model_kn.predict(x_val)
acc_kn=accuracy_score(y_val,kn_pred)
print(acc_kn)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
mod_nb= GaussianNB()
mod_svm = SVC(kernel='linear')
```

```python
gnb = mod_nb.fit(x_train,y_train) 
svm = mod_svm.fit(x_train,y_train) 
gnb_pred=gnb.predict(x_val)
svm_pred=svm.predict(x_val)
acc_gnb=accuracy_score(y_val,gnb_pred)
acc_svm=accuracy_score(y_val,svm_pred)
```
